```haskell
cuadrado :: Int -> Int
cuadrado a = a ^ 2
suma :: Int -> Int -> Int
suma x y = x + y

sumaCuadrados :: Int -> Int -> Int
sumaCuadrados x y = suma (cuadrado x) (cuadrado y)

duplicar x = x * 2
dobleLista :: [Int] -> [Int]
dobleLista = map duplicar 
```

```haskell
-- pruebas
sumaCuadrados 3 4
dobleLista [4, 2, 3, 8]
```

La forma mas sencilla de implementar 'dobleLista' seria con lista intencional, para cada elemento contenido en la lista lo multiplicamos por 2.
dobleLista xs = [x*2 | x <- xs]

```haskell
clasificarNota :: Float -> String
clasificarNota nota
        | nota < 3 = "Reprobado"
        | nota < 4 = "En recuperacion"
        | otherwise = "Aprobado"

finobacci 0 = 0
finobacci 1 = 1
finobacci n = finobacci (n - 1) + finobacci (n - 2)
```
