```haskell
head [10,20,30,40]
tail [10,20,30,40]
take 3 [1..10]
drop 4 [1..10]
length [1,2,3,4,5]
```

[1..10] genera una lista del numero 1 al 10. Se relaciona con la evaluacion perezosa porque haskell primero evalua la expresion antes de generar la lista, de tal forma que no llega a generar todos los elementos. Ej, en take 3 [1..10] se pide el primer elemento, 1, despues el segundo y el tercero, ahi termina la generacion.

```haskell
primerElemento :: [a] -> a
primerElemento (x:_) = x

productoLista :: [Int] -> Int
productoLista [] = 1
productoLista xs = head xs * productoLista (tail xs)

eliminarNegativos :: [Int] -> [Int]
eliminarNegativos xs = [x | x <- xs, x >= 0]

invertir :: [a] -> [a]
invertir [] = []
invertir (h:t) = invertir t ++ [h] 
```

```haskell
primerElemento [5, 1, 2]
productoLista [3, 3, 3]
eliminarNegativos [-2, 0, 1, 3, -9, 44]
invertir "Hola mundo"
invertir [1, 2, 3, 4, 5]
```
