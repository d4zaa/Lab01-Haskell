Significa que el lenguaje no evalua alguna funcion o expresion hasta que sea realmente requerida

```haskell
doble :: Int-> Int
doble x = x * 2

esMayorEdad :: Int -> Bool
esMayorEdad x
        | x >= 18   = True
        | otherwise = False

areaCirculo :: Float -> Float
areaCirculo r = pi * r^2
```
